```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(15, 13)}, font_scale=1)
```

```python
for col in df.columns:
    percent_null=np.mean((df[col]).isnull())
    print('{} --> {} %'.format(col,round(percent_null*100)))
```

```python
df.drop(['description','cast','show_id'], axis=1 ,inplace=True)
```

```python
df.dropna(subset=['date_added','rating','duration'],inplace=True)
```

```python
df['director'].fillna('Unknown', inplace=True)
```

```python
df['country'].fillna('other', inplace=True)
df['Country'] = df['country'].apply(lambda x: x.split(",")[0])
```

```python
df.isnull().sum()
```

```python
df['date_added']=pd.to_datetime(df['date_added'])
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
df['day_added']=df['date_added'].dt.day
```

```python
ratings_ages = {
    'TV-PG': 'Older Kids',
    'TV-MA': 'Adults',
    'TV-Y7-FV': 'Older Kids',
    'TV-Y7': 'Older Kids',
    'TV-14': 'Teens',
    'R': 'Adults',
    'TV-Y': 'Kids',
    'NR': 'Adults',
    'PG-13': 'Teens',
    'TV-G': 'Kids',
    'PG': 'Older Kids',
    'G': 'Kids',
    'UR': 'Adults',
    'NC-17': 'Adults'
}
df['target_ages'] = df['rating'].replace(ratings_ages)
df.head()
```

```python
df['season_count'] = df.apply(lambda x : x['duration'].split(" ")[0] if "Season" in x['duration'] else "", axis = 1)
df['duration'] = df.apply(lambda x : x['duration'].split(" ")[0] if "Season" not in x['duration'] else "", axis = 1)
df
```

```python
def map_months(x):
    if x in [12, 1, 2]:
        return 'Winter'
    elif x in [3, 4, 5]:
        return 'Spring'
    elif x in [6, 7, 8]:
        return 'Summer'
    elif x in [9, 10, 11]:
        return 'Autumn'
```

```python
movies['month_added'].apply(map_months)
```

```python
tv_show['month_added'].apply(map_months)
```

```python
plt.subplot(1,2,1)
sns.countplot(x=movies['month_added'].apply(map_months),data=movies)
plt.subplot(1,2,2)
sns.countplot(x=tv_show['month_added'].apply(map_months),data=tv_show)
```
